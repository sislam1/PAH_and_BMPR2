```python
import pandas as pd
import matplotlib.pyplot as plt
import re

df = pd.read_csv('chr2_variants.bed', sep='\t', header=None, names=['Chromosome', 'Start', 'End', 'AF_String', 'Column1', 'Column2', 'Column3'])
```

```python
df
```

```python
# Extract the allele frequency values from the "AF" entry
# Use regular expressions to find the "AF" value in the AF_String column
df['Allele_Frequency'] = df['AF_String'].str.extract(r'AF=(\d+\.\d+)').astype(float)

# Plot the distribution of allele frequencies
plt.hist(df['Allele_Frequency'], bins=20)  # Adjust the number of bins as needed
plt.xlabel('Allele Frequency')
plt.ylabel('Count')
plt.title('Distribution of Allele Frequencies')
plt.show()
```
