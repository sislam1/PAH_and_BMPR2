```python
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
# Replace 'data.xml' with the path to your XML file
tree = ET.parse('snp_result.xml')
root = tree.getroot()
```

```python
data_list = []

for element in root:
    data_dict = {}
    for child_element in element:
        data_dict[child_element.tag] = child_element.text
    data_list.append(data_dict)
```

```python
df = pd.DataFrame(data_list)
df
```
