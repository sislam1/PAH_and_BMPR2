```python
# library imports
import pandas as pd
```

```python
# read clinVar results
df = pd.read_table("clinvar_result.txt")
df
```

```python
# unique conditions
df['Condition(s)'].unique()
```

```python
search_string = "Pulmonary arterial hypertension"
count_occurrences = [0, 0, 0]
# Count the occurrences of the search string in the "Conditions" column (case insensitive)
count_occurrences[0] = df['Condition(s)'].str.contains(search_string, case=False).sum()
print(count_occurrences)
```

```python
search_string = "Pulmonary hypertension"

# Count the occurrences of the search string in the "Conditions" column (case insensitive)
count_occurrences[1] = df['Condition(s)'].str.contains(search_string, case=False).sum()
print(count_occurrences)
```

```python
search_string = "Pulmonary"

# Count the occurrences of the search string in the "Conditions" column (case insensitive)
count_occurrences[2] = df['Condition(s)'].str.contains(search_string, case=False).sum()
print(count_occurrences)
```

```python
import matplotlib.pyplot as plt
```

```python
diseases = ['Pulmonary arterial hypertension', 'Pulmonary hypertension', 'Pulmonary']
plt.bar(diseases, count_occurrences)
plt.xlabel('Search strings')
plt.ylabel('Count')
plt.title('Count for different strings')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()  # Adjusts the layout to prevent clipping of labels
plt.show()
```
